```python
import numpy as np
import matplotlib.pyplot as plt

# A(0,1), B(1,1), C(2,1)
# D(0,0), E(1,0), F(2,0)
x = np.array([[0, 1, 2], [0, 1, 2]])
y = np.array([[0, 0, 0], [1, 1, 1]])
# plt.plot(x, y, color='red', marker='.', linestyle='')
plt.plot(x, y, marker='.', markersize=10, linestyle='-.')
plt.grid(True)
plt.show()
```

X的每一行都一样，Y的每一列都一样

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2])
y = np.array([0, 1])

X, Y = np.meshgrid(x, y)
print(X)
print(Y)

plt.plot(X, Y, color='red', marker='.', linestyle='')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,1000,20)
y = np.linspace(0,500,20)

X, Y = np.meshgrid(x, y)
print(np.shape(X))
plt.plot(X, Y, color='red', marker='.', linestyle='')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X, Y = np.mgrid[0:1000:20j, 0:500:20j]
print(np.shape(X))
plt.plot(X, Y, color='red', marker='.', linestyle='')
plt.grid(True)
plt.show()
```
